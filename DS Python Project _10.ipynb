```python
# copy the data
copyy_df=df.copy()
copyy_df
```

```python
# Convert text to uppercase
copyy_df.columns = copyy_df.columns.str.upper()
copyy_df['TRANSACTION ID'] = copyy_df['TRANSACTION ID'].str.upper()
copyy_df['ITEM'] = copyy_df['ITEM'].str.upper()
copyy_df['PAYMENT METHOD'] = copyy_df['PAYMENT METHOD'].str.upper()
copyy_df['LOCATION'] = copyy_df['LOCATION'].str.upper()
copyy_df
```

```python
# Convert the 'DATE' column to datetime
copyy_df['TRANSACTION DATE'] = pd.to_datetime(copyy_df['TRANSACTION DATE'], errors='coerce')
```

```python
# DROP DATE COLUMNS
copyy_df.dropna(subset=['TRANSACTION DATE'], inplace=True)
copyy_df
```

```python
# Convert the 'QUANTITY', 'PRICE PER UNIT', 'TOTAL SPENT' columns to numeric
copyy_df['QUANTITY'] = pd.to_numeric(copyy_df['QUANTITY'], errors='coerce')
copyy_df['PRICE PER UNIT'] = pd.to_numeric(copyy_df['PRICE PER UNIT'], errors='coerce')
copyy_df['TOTAL SPENT'] = pd.to_numeric(copyy_df['TOTAL SPENT'], errors='coerce')
copyy_df
```

```python
df.dtypes
```

```python
# fill missing values with median because it is a numerical value & it doesn't get affected by outliers
copyy_df.fillna({'QUANTITY': copyy_df['QUANTITY'].median()}, inplace=True)
copyy_df.fillna({'PRICE PER UNIT': copyy_df['PRICE PER UNIT'].median()}, inplace=True)
copyy_df['TOTAL SPENT'] = copyy_df['TOTAL SPENT'].fillna(copyy_df['QUANTITY'] * copyy_df['PRICE PER UNIT'])
copyy_df
```

```python
# fill missing values with 'UNKNOWN'
copyy_df['ITEM'] = copyy_df['ITEM'].fillna('UNKNOWN')
copyy_df['PAYMENT METHOD'] = copyy_df['PAYMENT METHOD'].fillna('UNKNOWN')
copyy_df['LOCATION'] = copyy_df['LOCATION'].fillna('UNKNOWN')
copyy_df
```

```python
Q1 = copyy_df['TOTAL SPENT'].quantile(0.25)
Q3 = copyy_df['TOTAL SPENT'].quantile(0.75)
IQR = Q3 - Q1
Q1 , Q3 , IQR
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
lower, upper
```

```python
# identify outliers
outliers = copyy_df[
    (copyy_df['TOTAL SPENT'] < lower) |
    (copyy_df['TOTAL SPENT'] > upper)
]
outliers
```

```python
# cap outliers
copyy_df['TOTAL SPENT'] = copyy_df['TOTAL SPENT'].clip(lower, upper)
copyy_df
```

```python
# Boxplot after capping
plt.boxplot(copyy_df['TOTAL SPENT'])
plt.title('Total Spent After Capping')
plt.show()
# there is no outliers
```

```python
# Storing int & float
df_numeric = copyy_df.select_dtypes(include=['int64', 'float64'])
```

```python
# checking for nan cols
df_numeric = df_numeric.dropna(axis=1, how='all')
```

```python
df_numeric.shape
df_numeric.head()
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# fit & transform: calc. mean & std to every col , StandardScaler : standardize the numeric data
scaler = StandardScaler() 
# ÙƒÙ„ Ø§Ù„Ø¹ÙˆØ§Ù…ÙŠØ¯ Ù‚Ø¯ Ø¨Ø¹Ø¶ Ù„Ø§Ù† Ø§Ù„ key means Ø¨ÙŠØ¹Ù…Ù„ Ù…Ø³Ø§ÙÙ‡
scaled_data = scaler.fit_transform(df_numeric)  

scaled_df = pd.DataFrame(scaled_data, columns=df_numeric.columns)  #  make all col.s names the same

scaled_df.head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
# fit : mean , predict : every element in its cluster
clusters = kmeans.fit_predict(scaled_data)  

copyy_df['Cluster'] = clusters
copyy_df.head()
```

```python
inertia = [] 

for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(scaled_data)   # train the data 
    inertia.append(km.inertia_)

# dist. beteween centroids & other points
plt.figure(figsize=(8,5))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```
